```julia
using Interact
```

```julia
using ODE
```

```julia
using Plots
```

```julia
function LorenzEqs(a,b,c)
    f = function(t,r)
        (x, y, z) = r
        dx_dt = a*(y - x)
        dy_dt = x*(b - z) - y
        dz_dt = x*y - c*z
        [dx_dt; dy_dt; dz_dt]
    end
    return f
end
```

```julia
function drawAttractor(dT, T, a, b, c)
    t  = collect(0:float(dT):float(T))
    r0 = [0.1; 0.0; 0.0]
    f = LorenzEqs(a,b,c)
    (t, pos) = ode23(f, r0, t)
    x = map(v -> v[1], pos)
    y = map(v -> v[2], pos)
    z = map(v -> v[3], pos)
    plot3d(x,y,z)
end
```

```julia
@manipulate for dT in 0.001:0.001:0.01, T in 10.0:250.0, 
    a in 1:20, b in 10.0:1.0:50.0, c in 1.0:0.1:5.0
    drawAttractor(dT,T,a,b,c)
end
```
