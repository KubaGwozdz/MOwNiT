```julia
# mnożenie macierzy - wersja naiwna
function naive_multiplication(A,B)
C=zeros(Float64,size(A,1),size(B,2))
  for i=1:size(A,1)
    for j=1:size(B,2)
        for k=1:size(A,2)
            C[i,j]=C[i,j]+A[i,k]*B[k,j]
        end
    end
end
C
end
```

```julia
Asmall=[[1.0 0.0]; [0.0 1.0]]
Bsmall=Asmall
```

```julia
#kompilacja
naive_multiplication(Asmall,Bsmall)
```

```julia
#kompilacja funkcji BLASowej do mnożenia macierzy
#https://docs.julialang.org/en/stable/stdlib/linalg/#BLAS-Functions-1
Asmall*Bsmall
```

```julia
# poprawiona funkcja korzytająca z powyższego oraz z faktu, że
#można zmieniać kolejność operacji dodawania (a co za tym idzie kolejnosc petli).
function better_multiplication( A,B )
C=zeros(Float64,size(A,1),size(B,2))
  for j=1:size(B,2)
    for k=1:size(A,2)
        for i=1:size(A,1)
            C[i,j]=C[i,j]+A[i,k]*B[k,j]
        end
    end
end
C
end
```

```julia
better_multiplication(Asmall, Bsmall)
```

```julia
using DataFrames
using Statistics

Times = DataFrame(Size =[], BlasTime=[], BlasSD=[] ,NaiveTime=[], NaiveSD=[], BetterTime=[], BetterSD=[])

for it in 100:100:900
    matA = rand(it, it)
    matB = rand(it, it)
    Try = DataFrame(Blas=[], Naive=[], Better=[])
    for attempt in 1:10
        push!(Try,[(@elapsed (matA*matB))
        (@elapsed naive_multiplication(matA,matB))
        (@elapsed better_multiplication(matA,matB))])
    end
    push!(Times,[it,mean(Try[:Blas]), std(Try[:Blas]), mean(Try[:Naive]), std(Try[:Naive]), mean(Try[:Better]), std(Try[:Better])])
end
```

```julia
using Plots
plot(Times[:Size], [Times[:BlasTime] Times[:NaiveTime] Times[:BetterTime]]
,yerror = [Times[:BlasSD] Times[:NaiveSD] Times[:BetterSD]]
, title = "Time(Size)", xlabel = "Size", ylabel="Time[s]")
```
