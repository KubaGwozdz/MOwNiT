# Zadanie 1

```julia
# produkcja sztucznego sygnału.
#Wybieram częstotliwość próbkowania
Fs = 1024;  

# generuje wektor punktów czasowych w przedziale [0 1] co 1/Fs
t = 0:1/(Fs-1):1; 
```

```julia
x = sin.(2*pi*t*200) + 2* sin.(2*pi*t*400) 
```

```julia
using Plots
plot(t,x)
```

```julia
using FFTW
y = fft(x)
```

```julia
sticks((abs.(y)))
```

widzimy na wykresie częstotliwości 200 i 400

# Zadanie 2

```julia
xc = 2*cos.(2*pi*t*400) 
r = []
for i in 0:1/(Fs-1):1
    push!(r,rand(Float64))
end
xc = r+xc
```

```julia
plot(t,xc)
```

```julia
yc = fft(xc)
```

```julia
sticks(abs.(yc))
```

zgadza się dla 400

```julia
res = abs.(yc)

for i in 1:1:Fs
    if(res[i]) < 50
        res[i] = 0
    end
end
res
```

```julia
sticks(res)
```

```julia
clean = ifft(res)
```

```julia
plot(t, abs.(clean), label = "clean")
```

```julia
plot!(t, xc, label = "original")
```

```julia
plot!(t, abs.(clean), label = "clean")
```
