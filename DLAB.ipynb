```julia
@manipulate for n in 100:100:4000, f in 0:0.1:10
    f=10-f
    x = zeros(Int,n)
    y = zeros(Int,n)
    for i in 2:1:n
        if(x[i-1]<0 || y[i-1]<0)
            s=1
        else
            s=3
        end
        r = Int(floor(rand(TruncatedNormal(s, f, 0, 4))))
        if r==0
            x[i] = x[i - 1] + 1
            y[i] = y[i - 1] 
        elseif r==1
            x[i] = x[i - 1]
            y[i] = y[i - 1] + 1
        elseif r==2
            x[i] = x[i - 1] - 1
            y[i] = y[i - 1] 
        else
            x[i] = x[i - 1]
            y[i] = y[i - 1] -1
        end
    end
    scatter(x,y)
end
```
